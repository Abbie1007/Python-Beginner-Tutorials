```python
import cv2
import random

# 載入模型
face_cascade = cv2.CascadeClassifier(r'C:\Users\abbie\OneDrive\桌面\碩士課程\金融科技\haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(r'C:\Users\abbie\OneDrive\桌面\碩士課程\金融科技\haarcascade_eye.xml')

# 檢查模型是否載入成功
if face_cascade.empty():
    print("❌ 臉部模型載入失敗")
if eye_cascade.empty():
    print("❌ 眼睛模型載入失敗")

# 讀取圖片
img = cv2.imread(r'C:\Users\abbie\OneDrive\圖片\images.jpg')
if img is None:
    print("❌ 圖片讀取失敗")

# 灰階轉換
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 開始辨識
font = cv2.FONT_HERSHEY_SIMPLEX
faces = face_cascade.detectMultiScale(gray, 1.3, 5) 
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    cv2.putText(img, str(random.randrange(20, 30)), (x + (w // 2) - 18, y - 10), font, 1, (14, 201, 255), 2) 
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
